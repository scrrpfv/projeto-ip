# Visuzalização da dependência externa de petróleo 


## A dependência externa de energia de um país refere-se à necessidade desse país de importar energia de outras fontes para suprir suas necessidades energéticas. 

## A partir da análise do gráfico podemos inferir que o brasil apenas se tornou autosuficiente energeticamente no últimos 20 anos onde o aumento da energia primária junto com a mudança para uma dependência externa negativa sugere uma mudança positiva na dinâmica energética do país, indicando possivelmente uma maior segurança energética


```python
df.autoplot(0, tamanho=(10,8))
```

## Evidenciando os últimos 20 anos deixando mais evidente a recente mudança no paradigma da dependência externa


```python
colunas = df[0].columns
colunas_selecionadas = [colunas[0], colunas[3], colunas[4]]

ax = df[0][colunas_selecionadas][df[0].index > 2000].plot(title='EVOLUÇÃO DA DEPENDÊNCIA EXTERNA DE ENERGIA', figsize=(10,8), marker='o', grid=True)
ax.legend(loc='center right')
```

# Visualização da evolução da dependência externa de petróleo

## Podemos correlacionar a descoberta de grandes reservas de petróleo na camada pré-sal por volta de 2006 com a mudança de paradgima da depêndencias energeticas no Brasil.

##  O Brasil era altamente dependente das importações de petróleo para atender à sua demanda interna. No entanto, a descoberta das reservas de pré-sal marcou um ponto de virada significativo. Isso permitiu ao Brasil aumentar sua produção de petróleo e, consequentemente, reduzir sua dependência de importações. Assim, contribuindo para a redução da dependência de importações de petróleo e impulsionando a autonomia energética nacional.
