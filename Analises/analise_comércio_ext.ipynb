# Visuzalização da dependência externa de petróleo


## A dependência externa de energia de um país refere-se à necessidade desse país de importar energia de outras fontes para suprir suas necessidades energéticas.

## A partir da análise do gráfico podemos correlacionar a descoberta de grandes reservas de petróleo na camada pré-sal por volta de 2006 como principal agente na  mudança de  paradgima da depêndencias energeticas no Brasil.



```python
df.autoplot(0, tamanho=(10,8))
```

## Evidenciando os últimos 20 anos da mudança no aumento da produção primária de energia

```python
colunas = df[0].columns
colunas_selecionadas = [colunas[0], colunas[3], colunas[4]]

ax = df[0][colunas_selecionadas][df[0].index > pd.to_datetime(2000, format="%Y")].plot(title='EVOLUÇÃO DA DEPENDÊNCIA EXTERNA DE ENERGIA', figsize=(10,8), marker='o', grid=True)
ax.legend(loc='center right')
```

# Visualização da evolução da dependência externa de petróleo

## O Brasil era altamente dependente das importações de petróleo para atender à sua demanda interna. No entanto, a descoberta das reservas de pré-sal marcou um ponto de virada significativo. Isso permitiu ao Brasil aumentar sua produção de petróleo e, consequentemente, reduzir sua dependência de importações.

# Vizualização das importações de energia

## A partir da análise, iremos destacar a época do milagre econômico no Brasil, seguido pelo impacto da crise do petróleo nas importações de energia do país.

## O período de 1968 a 1973 ficou conhecido no Brasil como o "milagre econômico",  por uma redução gradual da inflação e uma melhora na balança comercial, com um aumento nas exportações em relação às importações.

## Em 1979 uma nova decisão da Opep traria o segundo choque do petróleo. Por decisão desse grupo de produtores, o preço do barril saiu de aproximadamente US$13,90 para US$35,69 em 1980.

## Essa nova onda da crise do petróleo fica efidente na diminuição das importações do pais que já estava bastante endividado.



### fontes:https://www.politize.com.br/decada-perdida-e-america-latina/

```python
df.autoplot(2)
```

# Evidenciando o período entre os anos 1970 - 2000

```python
colunas = df[2].columns
colunas_selecionadas = [colunas[0], colunas[8], colunas[19]]
ax = df[2][colunas_selecionadas][df[2].index < pd.to_datetime(2001, format="%Y")].plot(title='IMPORTAÇÕES DE ENERGIA', figsize=(10,8), marker='o', grid=True)
ax.legend(loc='center right')
```

# Vizualização da evolução das exportações e/ou importações líquidas

## O gráfico apresenta um balanço geral das importações e exportações. Nele, valores negativos indicam um aumento das exportações, enquanto valores positivos representam um aumento das importações.

## Quando as exportações superam as importações, temos um superávit comercial. Isso significa que o país está vendendo mais para o exterior do que comprando. Por outro lado, quando as importações são maiores que as exportações, ocorre um déficit comercial.
