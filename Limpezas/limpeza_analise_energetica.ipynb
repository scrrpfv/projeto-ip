# Separação das diferentes tabelas dentro da análise energética

```python
bigtable = pd.read_excel('../databases/Dados EPE/Análise Energética e Dados Agregados.xlsx', usecols='A:BB')
```

```python
dictd = {
    'Produção_de_Energia_Primária': bigtable[1:16].transpose().reset_index(drop=True),
    'Produção_de_Energia_Primária_%': bigtable[23:38].transpose().reset_index(drop=True),
    'Oferta_Interna_de_Energia': bigtable[44:58].transpose().reset_index(drop=True),
    'Oferta_Interna_de_Energia_%': bigtable[67:81].transpose().reset_index(drop=True),
    'Consumo_Final_por_Fonte': bigtable[89:115].transpose().reset_index(drop=True),
    'Consumo_Final_por_Fonte_%': bigtable[121:147].transpose().reset_index(drop=True),
    'Consumo_Final_por_Setor': bigtable[153:179].transpose().reset_index(drop=True),
    'Consumo_Final_por_Setor_%': bigtable[193:219].transpose().reset_index(drop=True),
    'Consumo_Final_Energético_por_Fonte': bigtable[225:249].transpose().reset_index(drop=True),
    'Consumo_Final_não_Energético_por_Fonte': bigtable[255:265].transpose().reset_index(drop=True),
    'Dependência_Externa_de_Energia': bigtable[271:285].transpose().reset_index(drop=True),
    'Composição_Setorial_do_Consumo_de_Derivados_de_Petróleo': bigtable[295:313].transpose().reset_index(drop=True),
    'Composição_Setorial_do_Consumo_de_Eletricidade': bigtable[321:331].transpose().reset_index(drop=True),
    'Composição_Setorial_do_Consumo_de_Carvão_Vapor': bigtable[337:349].transpose().reset_index(drop=True),
    'Composição_Setorial_do_Consumo_Final_de_Biomassa': bigtable[355:376].transpose().reset_index(drop=True),
    'Oferta_Interna_de_Energia': bigtable[384:393].transpose().reset_index(drop=True),
    'Oferta_Interna_de_Energia_%': bigtable[401:410].transpose().reset_index(drop=True),
}
```

```python
dict_analise_energetica = {}
for i in dictd:
    df = dictd[i]

    # Trocar os headers
    headers = pd.Series([str(v).strip() for v in df.iloc[0].values])
    headers[0] = 'ANO'
    df.columns = headers
    df.drop(index=0, axis=0, inplace=True)

    # Excluir colunas nulas
    df.dropna(axis='columns', how='all', inplace=True)
    
    # Trocar os valores dos anos de float para int
    df['ANO'] = df['ANO'].map(lambda x: int(x))

    dict_analise_energetica[i] = df
```
